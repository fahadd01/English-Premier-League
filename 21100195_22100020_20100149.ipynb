FEATURE REDUCTION 

FILTER METHOD

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(Y)
pleh = playing_stat.fillna(0.0)
pleh = pleh.assign(FTR= integer_encoded)
cor = pleh.corr()
cor_target = abs(cor["FTR"])
cor_target = cor_target.sort_values(ascending=False)
relevant_features = cor_target[abs(cor_target)>0.1]
relevant_features
```

```python
import seaborn as sns
col = list(relevant_features.index)
playint_stat_relevant = pleh[col].copy()
sns.heatmap(playint_stat_relevant.corr())
```

```python
corr_matrix = playint_stat_relevant.corr().abs()
sol = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
                  .stack()
                  .sort_values(ascending=False))
drop = sol[abs(sol)>0.5]
drop
```

```python
playint_stat_relevant = playint_stat_relevant.drop('FTR', axis=1)
playint_stat_relevant = playint_stat_relevant.drop('HTHG', axis=1)
playint_stat_relevant = playint_stat_relevant.drop('HTAG', axis=1)
playint_stat_relevant = playint_stat_relevant.drop('FTAG', axis=1)
playint_stat_relevant = playint_stat_relevant.drop('FTHG', axis=1)
```

EMBEDDED METHOD

```python
from sklearn.linear_model import LassoCV
reg = LassoCV()
YY = pleh['FTR']
reg.fit(playint_stat_relevant, YY)
coef = pd.Series(reg.coef_, index = playint_stat_relevant.columns)
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
imp_coef = coef.sort_values()
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Lasso Model")
```

K NEAREST NEIGHBOUR CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

def KNN_classifier(train_X,train_Y,test_X,test_Y):
    neigh = KNeighborsClassifier(n_neighbors=162)
    neigh.fit(train_X, train_Y)
    trainAccuracy = neigh.score(train_X,train_Y)
    yhat = neigh.predict(test_X)
    print('Predictions for test data : \n\n', yhat)
    print('\nAccuracy : ', accuracy_score(test_Y,yhat))
    print('\nMacro F1 score : ', f1_score(test_Y,yhat, average = 'macro'))
    print('\nConfusion Matrix : \n')
    print(confusion_matrix(test_Y,yhat))
KNN_classifier(train_X,train_Y,test_X,test_Y)
```

```python
from sklearn.linear_model import LogisticRegression
def logReg(train_X,train_Y,test_X,test_Y):
    clf = LogisticRegression(random_state=0, multi_class = 'ovr').fit(train_X, train_Y)
    yhat = clf.predict(test_X)
    print('Predictions for test data : \n\n', yhat)
    print('\nAccuracy : ', accuracy_score(test_Y,yhat))
    print('\nMacro F1 score : ', f1_score(test_Y,yhat, average = 'macro'))
    print('\nConfusion Matrix : \n')
    print(confusion_matrix(test_Y,yhat))
logReg(train_X,train_Y,test_X,test_Y)
```
